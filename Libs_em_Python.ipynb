```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas["estado"].fillna('MS', inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas.head()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
valor_errado = vendas.query("Preço > Preço_com_frete")
valor_correto = vendas.query("Preço < Preço_com_frete")
```

```python
valor_correto.head()
```

```python
valor_correto.describe()
```

```python
valor_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(valor_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
valor_correto.groupby('Data').idcompra.nunique().reset_index()
```

```python
valor_correto['Mes'] = valor_correto['Data'].dt.month
valor_correto.head()
```

```python
valor_correto.groupby('Mes').idcompra.nunique()
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head().reset_index()
```

```python
valor_correto.head()
```

```python
round(cliente[['idade','renda']].describe().reset_index(),2)
```

```python
cliente.info()
```

```python
cliente= cliente.astype({'renda':'float'})
```

```python
cliente.info()
```

```python
vendas_cliente = valor_correto.merge(cliente, how= 'left', on= 'cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending= False).reset_index(),2)
agg_idcanal_renda.head()
```

```python
agg_idade_bandeira= round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idade_bandeira.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')

plt.ylabel('Média de idade')

plt.title('Idade média por bandeira', loc='center', fontsize=18, fontweight='bold', color='black')
```

```python
fig2= plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='red')

plt.ylabel('Renda média por canal de venda')

plt.title('Compras por canal de vendas', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
agg_dpto_preco = round(valor_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dpto_preco
```

```python
fig = plt.figure(figsize=(30,20))

plt.bar(agg_dpto_preco['Nome_Departamento'], agg_dpto_preco['Preço_com_frete'], color='green')

plt.ylabel('Preço Médio')

plt.title('Preço médio por departamento', loc='center', fontsize=18, fontweight='bold', color='black')
```
